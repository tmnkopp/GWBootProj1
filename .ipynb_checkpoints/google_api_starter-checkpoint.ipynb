```python
# pip install --upgrade google-api-python-client

import os 
import pandas as pd
import googleapiclient.discovery
from config import api_key 
CONST_USE_CACHE = False 

  
def get_youtube(cache=False):  
    parts = 'status,contentDetails,snippet,statistics,id'
    if not cache:
        DEVELOPER_KEY = api_key  # <<<<---- PUT KEY HERE  
        os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1" 
        api_service_name = "youtube"
        api_version = "v3" 
        youtube = googleapiclient.discovery.build(
            api_service_name, api_version, developerKey = DEVELOPER_KEY)
        
        request = youtube.videos().list(
            part=parts,
            chart="mostPopular",
            locale="US",
            maxResults=100
        )

        json=request.execute()  
        df=pd.DataFrame(json['items']) 
        df.to_json('data/youtube.json')
    json=pd.read_json('data/youtube.json')
    df=pd.DataFrame(json)        
    return df


def get_cats(cache=False):   
    if not cache:
        DEVELOPER_KEY = api_key  # <<<<---- PUT KEY HERE  
        os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1" 
        api_service_name = "youtube"
        api_version = "v3" 
        youtube = googleapiclient.discovery.build(
            api_service_name, api_version, developerKey = DEVELOPER_KEY)
        
        request = youtube.videoCategories().list(
            part="snippet,id",
            regionCode="US"
        )

        json=request.execute()  
        df=pd.DataFrame(json['items']) 
        df.to_json('data/youtubecats.json')
    json=pd.read_json('data/youtubecats.json')
    df=pd.DataFrame(json)        
    return df 

         
df=get_cats(CONST_USE_CACHE) 
print(df.info())
df
#df pd.json_normalize(df['snippet']).head(35)
```
