```python
# pip install --upgrade google-api-python-client

import os 
import pandas as pd
import googleapiclient.discovery
from config import api_key 
 

def get_api():
    DEVELOPER_KEY = api_key  # <<<<---- PUT KEY HERE  
    os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1" 
    api_service_name = "youtube"
    api_version = "v3" 
    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey = DEVELOPER_KEY) 
    request = youtube.videos().list(
        part="snippet,statistics,id",
        chart="mostPopular",
        locale="US",
        maxResults=100
    )
    return request.execute()

df_items = pd.DataFrame(get_api()['items'])
df_snippet = pd.json_normalize(df_items['snippet'])
df_statistics = pd.json_normalize(df_items['statistics'])
df = pd.merge(df_items , df_snippet , left_index=True, right_index=True)
df_youtube = pd.merge(df , df_statistics , left_index=True, right_index=True )
print(df_youtube.info())
```

```python
df_youtube.drop(
    ['snippet','statistics'],axis=1, inplace=True
    )
df_youtube.drop(
    [c for c in df.columns if 'thumbnails' in c]
    , axis=1 , inplace=True
    ) 
df_youtube.fillna('0', inplace=True) 

for col in [c for c in df_youtube.columns if 'Count' in c]:
    df_youtube[col]=df_youtube[col].astype(int)
    
print(df_youtube.info())
df_youtube.head(5)
```
